```python
from selenium import webdriver
import pandas as pd
import time
from bs4 import BeautifulSoup
import re
```

# For Class E

```python
driver = webdriver.Firefox()
url = 'https://www.autohome.com.cn/197/#pvareaid=3311281'
driver.get(url)
rank_dict = {}
rank_table = driver.find_element_by_class_name('rank-con').find_elements_by_css_selector('tbody>tr')
for i in range(len(rank_table)):
    rank_elem = rank_table[i].text.split()
    rank_dict[i] = {'rank':i+1, 'car_name':rank_elem[0].split('.')[1], 'score':rank_elem[1], 'Advantage':rank_elem[2]}
    driver.implicitly_wait(20)
    new_window = rank_table[i].find_element_by_css_selector('td>a').click()
    time.sleep(3)
    driver.switch_to.window(driver.window_handles[1])
    driver.implicitly_wait(20)
    driver.find_element_by_link_text('口碑').click()
    driver.switch_to.window(driver.window_handles[2])
    time.sleep(3)
    try:
        driver.find_element_by_class_name('first-close.close').click()
        time.sleep(1)
    except:
        time.sleep(1)
    page = 20
    purp_lst = []
    post_num = 0
    for p in range(page):
        try:
            driver.find_element_by_class_name('first-close.close').click()
            time.sleep(1)
        except:
            time.sleep(1)
        purp = driver.find_elements_by_class_name('choose-dl.border-b-no')
        for p in purp:
            purp_lst += p.text.split()[1:]
            post_num += 1
        driver.implicitly_wait(20)
        driver.find_element_by_link_text('下一页').click()
        time.sleep(2)
    rank_dict[i]['pruchase_purpose'] = purp_lst
    rank_dict[i]['total_post'] = post_num
    driver.close()
    driver.switch_to.window(driver.window_handles[1])
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(20)
rank_dict
```

```python
car_info = pd.DataFrame(rank_dict).T
car_info
```

```python
car_info.to_csv('Class_E.csv',encoding='utf_8_sig')
```
