```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 
```

```python
dfc = pd.read_csv('..\data\data_C.csv',index_col = 0,dtype = {'simple review':str,'review':str,})
```

```python
sum(pd.notnull(dfc['review'])) # 3062 lines with review
dfc_review = dfc[pd.notnull(dfc['review'])] 
dfc_review = dfc_review.reset_index(drop = True)
dfc_review
```

```python
lst = []
for i in range(0,len(dfc_review)):
    if '[最满意]' in dfc_review.review[i]:
        lst.append(i)
pos_review = []
for i in dfc_review.loc[lst,'review'].tolist():
    review = i.split('\n')
    index = review.index('[最满意]')
    pos_review.append(review[index+1])
```

```python
dfc_pos_review1 = dfc_review.loc[lst,]
dfc_pos_review1['pos_review'] = pos_review
```

```python
lst2 = []
for i in range(0,len(dfc_review)):
    if '优点：' in dfc_review.review[i]:
        lst2.append(i)
lst2 = lst2[2:]    #delete 552 and 661.have both '[最满意]' and'优点：'
pos_review = []
for i in dfc_review.loc[lst2,'review'].tolist():
    review = i.split('\n')
    index = review.index('优点：')
    pos_review.append(review[index+1])
```

```python
dfc_pos_review2 = dfc_review.loc[lst2,]
dfc_pos_review2['pos_review'] = pos_review
dfc_pos_review = pd.concat([dfc_pos_review1, dfc_pos_review2])
```

```python
dfc_pos_review.to_csv('..\data\data_C_positive_review.csv')
```
