```python
files = os.listdir()
```

```python
comb_data=pd.DataFrame()
for f in files:
    if f.endswith('_o.csv'):
        df = pd.read_csv(f,index_col=0)
        df['from'] = f.split('_')[0]
        df.drop_duplicates(keep='first',inplace=True)
        comb_data = pd.concat([comb_data,df],ignore_index=True)
```

```python
from pyecharts import Map, Geo
```

```python
map_table=pd.DataFrame()
for i in range(len(comb_data)):
  if comb_data.loc[i,'region'] not in ['其他','海外']:
    map_table = pd.concat([map_table,comb_data.iloc[i:i+1,]])
```

```python
brand_name = list(set(comb_data['from']))
for brand in brand_name:
  sub = map_table.loc[map_table['from']==brand]
  count = Counter(sub['region'].to_list())
  province_distribution = count
  provice=list(province_distribution.keys())
  values=list(province_distribution.values())
  map = Map("Geographical Distribution",brand, width=1200, height=600)
  map.add("", provice, values, visual_range=[0, 50],  maptype='china', is_visualmap=True,
      visual_text_color='#000')
  map.show_config()
  map.render(path=brand+'.html')
```
