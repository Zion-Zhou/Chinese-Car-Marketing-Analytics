```python
import pandas as pd
from collections import Counter
```

```python
data = pd.read_csv('interest data/奔驰_o.csv',index_col=0)
data.head()
```

```python
tag_lst = data['tag'].to_list()
standard_lst = []
for tags in tag_lst:
    ele = tags.split(', ')
    standard_lst = standard_lst + ele
```

```python
counter1 = Counter(standard_lst)
counter1
```

```python
counter2 = {}
for tag,num in counter1.items():
    if num > 10 and tag != "奔驰" and  tag !='':
        counter2[tag]=num
stat = pd.DataFrame(data = counter2,index=['Freq']).T
stat.to_csv('benz-wordcloud.csv',  encoding = 'utf_8_sig')
```

```python
cloud_text=[]
for name, freq in  counter2.items():
    cloud_text = cloud_text + [name]*freq
```

```python
counter3 = pd.DataFrame(cloud_text)
counter3.to_csv('benz-wordcloud.txt', index = False, encoding = 'utf_8_sig')
```
